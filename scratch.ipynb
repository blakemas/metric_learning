Confirm that L12 norm scales like $\sqrt{pd}$

```python
import numpy as np
from scipy.linalg import orth
import matplotlib.pyplot as plt
l12 = []
for p in range(2, 100):
    U = orth(np.random.randn(p, 2))
    Ktrue = np.dot(U, U.T)
    l12.append(sum([np.linalg.norm(row) for row in Ktrue]))
plt.plot(l12)
plt.show()
print np.polyfit(np.log(range(2,100)), np.log(l12), 1)

l12 = []
for d in range(1, 10):
    U = orth(np.random.randn(100, d))
    Ktrue = np.dot(U, U.T)
    l12.append(sum([np.linalg.norm(row) for row in Ktrue]))
plt.plot(l12)
plt.show()
print np.polyfit(np.log(range(1,10)), np.log(l12), 1)
```
